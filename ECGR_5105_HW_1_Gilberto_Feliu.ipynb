```python
# ECGR 5105 Homework 1
# Gilberto Feliu

# Setup Code
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# Get data from CSV
url = 'https://raw.githubusercontent.com/HamedTabkhi/Intro-to-ML/b4cf4ac187d647a443dc011c4ee85fbb42e26096/Dataset/D3.csv'
df = pd.read_csv(url)
#print(df.head(10))

# Separate features and labels
X1 = df.values[:, 0]
X2 = df.values[:, 1]
X3 = df.values[:, 2]
Y = df.values[:, 3]
len_x1 = len(X1)
len_x2 = len(X2)
len_x3 = len(X3)
len_y = len(Y)
```

```python
# Function definitions
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)  # Number of training examples
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        predictions = X.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha / m) * X.transpose().dot(errors)
        theta -= sum_delta
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history

def compute_cost(X, y, theta):
    predictions = X.dot(theta)
    errors = np.subtract(predictions, y)
    sqrErrors = np.square(errors)
    J = 1 / (2 * len_x1) * np.sum(sqrErrors)
    return J
```

```python
# problem 1
# run linear regression separately for x1, x2, x3 (one at a time) with alpha between 0.01 and 0.1
# 1. report the linear model (theta values) for each variable
# 2. plot the fitted line and also plot loss vs iterations for each variable
# 3. figure out which variable gives the lowest cost for predicting Y
# 4. write about how the different learning rates changed the loss and how many iterations it took

# Linear regression for X1
X0 = np.ones((len_x1, 1))
X_1 = X1.reshape((len_x1, 1))
feature_mat = np.hstack((X0, X_1))
theta = np.zeros(2)
iterations = 1500
alpha = 0.01
theta, cost_history = gradient_descent(feature_mat, Y, theta, alpha, iterations)
print('For explanatory variable X1')
print('Final value of theta =', theta)
print('cost_history =', cost_history)
print('\n')

# Data Visualization
plt.scatter(feature_mat[:, 1], Y, color='red', marker='+', label='Training Data')
plt.plot(feature_mat[:, 1], feature_mat.dot(theta), color='green', label='Linear Regression')
# Plot customizations
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.title('Linear Regression Fit')
plt.legend()
# Show the plot
plt.show()

# Plot loss over iterations
plt.clf()
plt.plot(range(iterations), cost_history, color='blue')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Loss vs. Iterations')
plt.show()

# Linear regression for X2
X0 = np.ones((len_x2, 1))
X_2 = X2.reshape((len_x2, 1))
feature_mat = np.hstack((X0, X_2))
theta = np.zeros(2)
iterations = 1500
theta, cost_history = gradient_descent(feature_mat, Y, theta, alpha, iterations)
print('For explanatory variable X2')
print('Final value of theta =', theta)
print('cost_history =', cost_history)
print('\n')

# Data Visualization
plt.clf()
plt.scatter(feature_mat[:, 1], Y, color='red', marker='+', label='Training Data')
plt.plot(feature_mat[:, 1], feature_mat.dot(theta), color='green', label='Linear Regression')
# Plot customizations
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.title('Linear Regression Fit')
plt.legend()
# Show the plot
plt.show()

# Plot loss over iterations
plt.clf()
plt.plot(range(iterations), cost_history, color='blue')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Loss vs. Iterations')
plt.show()

# Linear regression for X3
X0 = np.ones((len_x3, 1))
X_3 = X3.reshape((len_x3, 1))
feature_mat = np.hstack((X0, X_3))
theta = np.zeros(2)
iterations = 1500
theta, cost_history = gradient_descent(feature_mat, Y, theta, alpha, iterations)
print('For explanatory variable X3')
print('Final value of theta =', theta)
print('cost_history =', cost_history)

# Data Visualization
plt.scatter(feature_mat[:, 1], Y, color='red', marker='+', label='Training Data')
plt.plot(feature_mat[:, 1], feature_mat.dot(theta), color='green', label='Linear Regression')
# Plot customizations
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.title('Linear Regression Fit')
plt.legend()
# Show the plot
plt.show()

# Plot loss over iterations
plt.clf()
plt.plot(range(iterations), cost_history, color='blue')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Loss vs. Iterations')
plt.show()
```
